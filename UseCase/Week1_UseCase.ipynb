```python
# Fill dataset with 0 instead of NaN

train_data.fillna(value=0,inplace=True)
test_data.fillna(value=0,inplace=True)
```

```python
# Creating different scatterplot between training and testing set for first 5 variables 

def plot_scatter(td1,td2,features):
    a=1
    plt.subplots(figsize=(20,20))
    for f in features:
        plt.subplot(5,5,a)
        plt.scatter(td1[f], td2[f], label=f)
        plt.title(f, fontsize=9)
        a+=1
    
feats = ['var_%s'%i for i in range(0,5)]
plot_scatter(train_data, test_data, feats)
```

```python
# Creating countplot and finding percentage of values with target 0 and 1

plot1 = sns.countplot(x=train_data["target"],palette='Set2')
```

```python
ones = train_data['target'].value_counts()[1]
print("There are {}% target with value 1".format(100*ones/train_data['target'].count()))
```

```python
# Creating distribution density plot for first 5 variables for training set

def plot_dist(td1,feat):
    
    fig, axes = plt.subplots(figsize=(20,24))
    f0 = train_data.loc[train_data['target']==0]
    f1 = train_data.loc[train_data['target']==1]
    a=1
    for f in feat:
        plt.subplot(5,5,a)
        sns.distplot(f0[f],hist=False,color='red',label="0")
        sns.distplot(f1[f],hist=False,label="1")
        a+=1
        plt.legend(loc="upper right")

features = ['var_%s'%i for i in range(0,5)]
plot_dist(train_data,features)
```

```python
# Plot showing mean values per train entry in the train set

plt.figure(figsize=(10,4))
features = train_data.columns.values

plt.title("Distribution of mean values per train entry in the train set")
sns.histplot(train_data[features].mean(axis=1), kde= True, color='black', bins=100, label="Train")

plt.legend()
plt.show()
```

```python
# Plot showing mean values per test entry in the test set

plt.figure(figsize=(10,4))
features = test_data.columns.values

plt.title("Distribution of mean values per test entry in the test set")
sns.histplot(test_data[features].mean(axis=1), kde= True, color='red', bins=100, label="Test")

plt.legend()
plt.show()
```

```python
# Plot showing distribution of standard deviation values in train and test sets

plt.figure(figsize=(10,4))

features1 = train_data.columns.values
features2 = test_data.columns.values

sns.histplot(train_data[features1].std(axis=0), kde= True, color='black', bins=100, label="Train")
sns.histplot(test_data[features2].std(axis=0), kde= True, color='red', bins=100, label="Test")

plt.title("Distribution of standard deviation values in train and test sets")
plt.legend()
plt.show()
```
